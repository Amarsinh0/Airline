```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Test_df=pd.read_excel("Test_set.xlsx")
Test_df.head()
```

```python
Train_df=pd.read_excel("Data_Train.xlsx")
Train_df.head()
```

```python
df=Train_df.append(Test_df)
df.head()
```

```python
df.info()
```

```python
df['date']=df['Date_of_Journey'].str.split('/').str[0]
df['month']=df['Date_of_Journey'].str.split('/').str[1]
df['year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df['Date']=df['date'].astype(int)
df['Month']=df['month'].astype(int)
df['Year']=df['year'].astype(int)
```

```python
df.info()
```

```python
df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'] = pd.to_datetime(df['Arrival_Time']).dt.time
df.head()
```

```python
df.isnull().sum()
```

```python
df['Arrival_Time']
```

```python
# Convert the 'Arrival_Time' column to string type
df['Arrival_Time'] = df['Arrival_Time'].astype(str)

# Split the 'Arrival_Time' column into hours and minutes
df['Arrival_hours'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]

```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
# Convert the 'Dep_Time' column to string type
df['Dep_Time'] = df['Dep_Time'].astype(str)

# Split the 'Dep_Time' column into hours and minutes
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1].astype(int)

# Drop the original 'Dep_Time' column
df.drop("Dep_Time", axis=1, inplace=True)
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
# Define the mapping dictionary
stop_mapping = {
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4,
    np.nan: np.nan  # If 'nan' values are represented as np.nan
    # 'nan': -1  # If 'nan' values should be mapped to a specific value, like -1
}

# Apply the mapping to the 'total_stops' column
df['Total_Stops'] = df['Total_Stops'].map(stop_mapping)
```

```python
df.head()
```

```python
df.drop('Route',axis=1,inplace=True)
df.head()
```

```python
def duration_to_minutes(duration_str):
    hours = 0
    minutes = 0
    if 'h' in duration_str:
        hours = int(duration_str.split('h')[0])
    if 'm' in duration_str:
        minutes = int(duration_str.split('m')[0].split()[-1])
    return hours * 60 + minutes

# Apply the function to the 'Duration' column
df['Duration_minutes'] = df['Duration'].apply(duration_to_minutes)
df['Duration_minutes']

```

```python
df.head()
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the 'Airline' column
df['Airline'] = label_encoder.fit_transform(df['Airline'])
df['Source'] = label_encoder.fit_transform(df['Source'])
df['Destination'] = label_encoder.fit_transform(df['Destination'])
df['Additional_Info'] = label_encoder.fit_transform(df['Additional_Info'])
```
